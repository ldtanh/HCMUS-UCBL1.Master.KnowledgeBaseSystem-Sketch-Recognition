```python
import os
os.chdir('TXT')
categories = os.listdir()
```

```python
def process_line(line, vector):
    attr=map(lambda x: int(x), list(line.strip()))
    vector.extend(list(attr))

def process_vector(file, vector):
    lines = file.read().splitlines()
    for line in lines[1:]:
        process_line(line, vector)
        
def process_file(file, vectors, targets, category):
    vector=[]
    process_vector(file, vector)
    vectors.append(vector)
    targets.append(category)
```

```python
vectors=[]
targets=[]
for num, category in enumerate(categories, start=1):
    os.chdir(category)
    
    filenames = os.listdir()
    for num, filename in enumerate(filenames, start=1):
        with open(filename) as file:
            process_file(file, vectors, targets, category)

    print('Read '+category)
    os.chdir('..')
```
